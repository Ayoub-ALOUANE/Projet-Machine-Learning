$$
\begin{aligned}
X &= \begin{bmatrix}

X_1^{(1)}X_2^{(1)} & X_1^{(1)}X_3^{(1)} & \cdots & X_{11}^{(1)}X_{12}^{(1)} &
1 \\
X_1^{(2)} & X_2^{(2)} & \cdots & X_{12}^{(2)} &
(X_1^{(2)})^2 & (X_2^{(2)})^2 & \cdots & (X_{12}^{(2)})^2 &
X_1^{(2)}X_2^{(2)} & X_1^{(2)}X_3^{(2)} & \cdots & X_{11}^{(2)}X_{12}^{(2)} &
1 \\
X_1^{(3)} & X_2^{(3)} & \cdots & X_{12}^{(3)} &
(X_1^{(3)})^2 & (X_2^{(3)})^2 & \cdots & (X_{12}^{(3)})^2 &
X_1^{(3)}X_2^{(3)} & X_1^{(3)}X_3^{(3)} & \cdots & X_{11}^{(3)}X_{12}^{(3)} &
1 \\
\vdots & \vdots & \ddots & \vdots &
\vdots & \vdots & \ddots & \vdots &
\vdots & \vdots & \ddots & \vdots &
\vdots \\

\end{bmatrix}
\end{aligned}
